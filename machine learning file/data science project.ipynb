```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("/content/etsy-mens-t-shirts-11-15-2017.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
test= data.dropna()
```

```python
test=test.reset_index(drop=True)
```

```python
test.isnull()
```

```python
gb_sellername = test.groupby('seller_name')['product_price'].sum()
```

```python
gb_sellername.describe()
```

```python
print(gb_sellername)
```

```python
gb_sellername = gb_sellername.sort_values(ascending=False)
```

```python
print(gb_sellername)
```

```python
a=gb_sellername[:10]
```

```python
a.plot(kind='bar')
```

```python
gb_sellername = test.groupby('seller_name')['product_price'].count()
```

```python
print(gb_sellername)
```

```python
gb_sellername = gb_sellername.sort_values(ascending=False)
print(gb_sellername)
```

```python
a=gb_sellername[:10]
plt.figure(figsize=(15,8))
a.plot(kind='bar',color='g')
```

```python
plt.figure(figsize=(15,8))
a.plot(kind='hist',color='r')
```
